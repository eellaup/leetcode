# [LeetCode 198. House Robber](https://leetcode.com/problems/house-robber/description/)

Time Complexity: O(n)

Space Complexity: O(1)

```python
class Solution:
    def rob(self, nums: List[int]) -> int:
        rob1, rob2 = 0, 0

        for n in nums:
            rob1,rob2 = rob2, max(rob1+n, rob2)
        
        return rob2
```
